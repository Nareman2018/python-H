## Welcome to the Python Course!

In this course, we will be exploring the fundamentals of Python programming. This course is designed for beginners and covers a wide range of topics to help you get started with Python. By the end of the course, you will have a strong foundation in Python and be ready to tackle more advanced topics.

### What You Will Need to Follow Along

#### Visual Studio Code (VS Code) with the Jupyter Extension Installed
VS Code is a powerful, lightweight code editor that supports various programming languages. The Jupyter extension for VS Code allows you to work with Jupyter Notebooks directly within the editor, providing a seamless development experience.

> **Note:** You are free to use whatever code editor you want based on your personal preference, however the tools mentioned here is what I am using and recommending to follow along.

#### What is Jupyter?
Jupyter is an open-source project that provides an interactive computing environment. A Jupyter Notebook is a web-based application that allows you to create and share documents containing live code, equations, visualizations, and narrative text. It's an excellent tool for data analysis, visualization, and interactive programming.

#### Poetry for Environment Management
Poetry is a dependency manager for Python that helps you manage your project's dependencies and virtual environments. It simplifies the process of setting up and maintaining a consistent development environment.
